##### Explanation:
* Firstly, load the Image in python.
* Then converted the image to grayscale.
* Then invert the image.
* Next apply Gaussian Blur.
* Finish it by douging and burning then converting the image to a Pencil Sketch.

##### IMPORTING THE LIBRARIES AND LOAD THE DATA

```python
#IMPORT LIBRARIES
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('tiger.jpg',0)
plt.imshow(img)
plt.show()
```

```python
#Color value just for own sake
print('Minimum color range',img.min())
print('Maximum color range',img.max())
```

### CONVERTING TO GRAYSCALE

```python
#Converting image to greyscale
plt.imshow(img,cmap='gray')
plt.title('Black and white imege')
plt.show()
```

### INVERTING THE IMAGE

```python
#Inverting the image
inv=255-img
plt.title('Inverted imege')
plt.imshow(inv,cmap='gray')
plt.show()
```

### GAUSSIAN BLUR

```python
#Applying Gaussian blur to smooth out the image
gblur=cv2.GaussianBlur(inv,ksize=(21,21),sigmaX=0,sigmaY=0)
plt.title('Image after Gaussian blur applied')
plt.imshow(gblur,cmap='gray')
plt.show()
```

### DOUGING AND BURNING

```python
#Dodging and burninG
dodge= lambda image,mask: cv2.divide(image,255-mask,scale=256)
blended=dodge(img,gblur)
plt.title('Pencil sketch of image')
plt.imshow(blended,cmap='gray')
plt.show()
```
